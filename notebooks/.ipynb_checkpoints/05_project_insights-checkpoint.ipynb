```python

import pandas as pd
import numpy as np
import os


import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib


from tqdm import tqdm
tqdm.pandas()
```

```python
data_path = "../data/processed/climate_nasa_preprocessed.csv"
df = pd.read_csv(data_path)


df = df[df['clean_text'].notna() & (df['clean_text'] != "")]
df.head()
```

```python
print("Number of comments:", df.shape[0])
print("Columns:", df.columns.tolist())
print("Missing values:\n", df.isna().sum())


plt.figure(figsize=(12,5))
sns.histplot(df['likesCount'], bins=30, kde=True)
plt.title("Distribution of Likes")
plt.show()

plt.figure(figsize=(12,5))
sns.histplot(df['commentsCount'], bins=30, kde=True)
plt.title("Distribution of Comments")
plt.show()
```

```python

clf = joblib.load("../src/models/sentiment_clf.pkl")
tfidf = joblib.load("../src/models/tfidf_vectorizer.pkl")


X_vec = tfidf.transform(df['clean_text'])


df['sentiment'] = clf.predict(X_vec)


sentiment_order = ['negative', 'neutral', 'positive']


plt.figure(figsize=(8,5))
sns.countplot(
    data=df,
    x='sentiment',
    order=sentiment_order
)
plt.title("Sentiment Distribution")
plt.show()


sentiment_counts = df['sentiment'].value_counts().reindex(sentiment_order, fill_value=0)
print(sentiment_counts)
```

```python

nmf_model = joblib.load("../src/models/nmf_topic_model.pkl")
tfidf_topic = joblib.load("../src/models/tfidf_vectorizer_topic.pkl")


tfidf_matrix = tfidf_topic.transform(df['clean_text'])
topic_values = nmf_model.transform(tfidf_matrix)
df['dominant_topic'] = topic_values.argmax(axis=1)


def print_top_words(model, feature_names, n_top_words=10):
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        print(f"Topic {topic_idx+1}: {' | '.join(top_features)}")

tfidf_feature_names = tfidf_topic.get_feature_names_out()
print_top_words(nmf_model, tfidf_feature_names)


plt.figure(figsize=(8,5))
sns.countplot(x='dominant_topic', data=df)
plt.title("Distribution of Dominant Topics")
plt.show()
```

```python

engagement = df.groupby('sentiment')[['likesCount','commentsCount']].mean()
print(engagement)


engagement.plot(kind='bar', figsize=(10,5), ylabel='Average Engagement')
plt.title("Average Likes & Comments per Sentiment")
plt.show()
```

```python
for topic_idx, topic in enumerate(nmf_model.components_):
    word_freq = {tfidf_feature_names[i]: topic[i] for i in topic.argsort()[:-50 - 1:-1]}
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
    
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"WordCloud for Topic {topic_idx+1}")
    plt.show()
```
