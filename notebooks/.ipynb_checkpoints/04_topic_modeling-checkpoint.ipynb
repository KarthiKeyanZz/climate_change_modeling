```python

import pandas as pd
import numpy as np
import os


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF


import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
preprocessed_path = "../data/processed/climate_nasa_preprocessed.csv"
df = pd.read_csv(preprocessed_path)


df = df[df['clean_text'].notna() & (df['clean_text'] != "")]
df.shape
```

```python
tfidf = TfidfVectorizer(max_features=5000, stop_words='english', ngram_range=(1,2))
tfidf_matrix = tfidf.fit_transform(df['clean_text'])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

```python

n_topics = 5

nmf_model = NMF(n_components=n_topics, random_state=42)
nmf_model.fit(tfidf_matrix)
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        top_weights = topic[top_features_ind]
        print(f"Topic {topic_idx+1}: {' | '.join(top_features)}")

tfidf_feature_names = tfidf.get_feature_names_out()
print_top_words(nmf_model, tfidf_feature_names, n_top_words=10)
```

```python
topic_values = nmf_model.transform(tfidf_matrix)
df['dominant_topic'] = topic_values.argmax(axis=1)

df[['clean_text', 'dominant_topic']].head(10)
```

```python
topic_values = nmf_model.transform(tfidf_matrix)
df['dominant_topic'] = topic_values.argmax(axis=1)

df[['clean_text', 'dominant_topic']].head(10)
```

```python
for topic_idx, topic in enumerate(nmf_model.components_):
    print(f"WordCloud for Topic {topic_idx+1}")
    wordcloud = WordCloud(width=800, height=400, background_color='white')
    word_freq = {tfidf_feature_names[i]: topic[i] for i in topic.argsort()[:-50 - 1:-1]}
    wordcloud.generate_from_frequencies(word_freq)
    
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
import joblib

os.makedirs("../src/models", exist_ok=True)

joblib.dump(nmf_model, "../src/models/nmf_topic_model.pkl")
joblib.dump(tfidf, "../src/models/tfidf_vectorizer_topic.pkl")
print("NMF model and vectorizer saved!")
```
