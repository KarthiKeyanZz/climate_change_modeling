```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from tqdm import tqdm
import datetime

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
sns.set_style('whitegrid')
```

```python
raw_data_path = "../data/raw/climate_nasa.csv"
df = pd.read_csv(raw_data_path)
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df[['date']].info()
```

```python
# Text column: drop or mark missing
print("Missing text values:", df['text'].isna().sum())

# Option 1: drop rows with missing text
df = df.dropna(subset=['text'])

# Comments count: fill missing with 0 (assume no replies)
df['commentsCount'] = df['commentsCount'].fillna(0)

# Confirm
df.info()
```

```python

print("Dataset shape:", df.shape)
print(df.describe())
print(df['likesCount'].value_counts().head())
```

```python

df['text_length'] = df['text'].apply(len)
plt.figure(figsize=(10,5))
sns.histplot(df['text_length'], bins=30, kde=True)
plt.title("Distribution of Comment Text Length")
plt.xlabel("Text Length")
plt.ylabel("Frequency")
plt.show()
```

```python

plt.figure(figsize=(10,5))
sns.histplot(df['likesCount'], bins=30, kde=True)
plt.title("Distribution of Likes Count")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['commentsCount'], bins=30, kde=True)
plt.title("Distribution of Comments Count")
plt.xlabel("Comments")
plt.ylabel("Frequency")
plt.show()
```

```python

all_text = " ".join(df['text'].tolist())


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)


plt.figure(figsize=(15,7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud of NASA Climate Comments")
plt.show()
```

```python

processed_path = "../data/processed/climate_nasa_cleaned.csv"
df.to_csv(processed_path, index=False)
print(f"Cleaned dataset saved to {processed_path}")
```
